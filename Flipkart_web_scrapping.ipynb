```python
import requests # for request data
from bs4 import BeautifulSoup # for data prasing libraries
import pandas as pd # for data manupulation
import numpy as np # for data manupulation
```

```python
def get_title(soup,d):
    
    try:
        title=soup.find_all('div',attrs={'class':'_4rR01T'})
        
        for i in range(1,24):
            title_string=title[i].text.strip()
            d['Title'].append(title_string) 
    except AttributeError:
        d['Title'].append(title_string='')

def get_price(soup,d):
    
    try:
        price=soup.find_all('div',attrs={'class':'_30jeq3 _1_WHN1'})
        for i in range(1,24):
            price_string=price[i].text.strip()
            d['Price'].append(price_string)
    except AttributeError:
        d['Price'].append(price_string='') 
        
def get_rating(soup,d):
    
    try:
        rating=soup.find_all('div',attrs={'class':'_3LWZlK'})
        for i in range(1,24):
            rating_string=rating[i].text.strip()
            d['Rating'].append(rating_string)
    
    except AttributeError:
        d['Rating'].append(rating_string='')    
   
        
def get_description(soup,d):
    
    try:
        description=soup.find_all('ul',attrs={'class':'_1xgFaf'})
        for i in range(1,24):
            description_string=description[i].text.strip()
            d['Description'].append(description_string)
    
    except AttributeError:
        
        d['Description'].append(description_string='')     
         
def get_offer(soup,d):
    
    try:
        offer=soup.find_all('div',attrs={'class':'_3Ay6Sb'})
        for i in range(1,24):
            offer_string=offer[i].text.strip()
            d['Offer'].append(offer_string)  
    
    except AttributeError:
        d['Offer'].append(offer_string='')                              
```

```python
if __name__=='__main__':

    HEADER=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36','Accept-Language': 'en-US, en;q=0.5'})
    
    d={'Title':[],'Price':[],'Rating':[],'Description':[],'Offer':[]}
    
    for i in range(2,20):
            
            ## Add URL
        URL ='https://www.flipkart.com/search?q=laptops&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=laptops&requestId=f0a883b3-ae8f-4d6a-a992-813bf2a7dc0d&page='+str(i)
            
            ## HTTP request
        R= requests.get(URL,headers=HEADER)
            
            # soup object containing all data
        soup=BeautifulSoup(R.content,'html.parser')
            
        
        get_title(soup,d)
        get_price(soup,d)
        get_rating(soup,d)
        get_description(soup,d)
        get_offer(soup,d)
        
       
    Flipkart_df=pd.DataFrame.from_dict(d)
    
    Flipkart_df.to_csv('Flipkart_data.csv',header=True,index=False)
```
